# ESTRUTURA DE REPETIÇÃO


##For##

* Executar comandos, enquanto percorre os itens de uma coleção (lista, tuplas, dicionários e sequência de números inteiros)

    >          For <variável> in <coleção>
    >          <comandos>

*Formas de repetição*

* Range
* Simples
* Enumerante

```python
# For Range
for contador in range(10):  # Range - cria um objeto do tipo range que cria uma sequência numérica
  print(contador)
```

```python
# For Range + Break
for contador in range(1000):
  print(contador)
  if contador == 7:
    break
```

```python
lista = ["Teste1", "Teste2", "Teste3"]
```

```python
# For Simples
for val in lista:   # Comando utilizado para percorrer uma lista inteira, sem necessitar que você saiba a última posição dessa lista
  print(val)
```

```python
# For com Enumerate
for i, val in enumerate(lista):   # Comando utilizado para enumerar os objetos da coleção, é necessário sempre indicar o índice (i)
  print(i, val)
```

##For - Operadores compostos##

* Muito utilizados em laços de repetição
* Código mais utilizado

```python
lista2 = [2, 3, 4]
```

```python
for val in lista2:
  val += 1
  print(val)
```

##While##

* Executar comandos enquanto uma condição é verdadeira

>          while<condição>:
                <comandos>

          
>          while<condição>:
                <comandos>
           else:
               <comandos>

```python
# While
val = 1
while (val < 7):
  val *= 2
  print(val)
```

```python
# While com else
val = 4
while (val < 10):
  val *= 2
else:
  print("Valor de val é: ", val)
```
