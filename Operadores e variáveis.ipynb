#Tabela da Verdade



And

```python
print(20 > 10 and 4 < 8)     # V and V - V
print(10 > 5 and 5 > 7)      # V and F - F
print(10 < 5 and 4 < 6)      # F and V - F
print(10 < 2 and 6 < 3)      # F and F - F
```

Or

```python
print(10 > 2 or 4 < 6)       # V or V - V
print(10 > 2 or 4 > 6)       # V or F - V
print(10 < 2 or 4 < 6)       # F or V - V
print(10 < 2 or 4 > 6)       # F or F - F
```

Not

```python
not 10 > 2      # V - F
```

```python
not 10 < 2      # F - V
```

#Variáves
Variáveis estão alocadas em espaços na memória

```python
x = 10    # Nome da variável é "x" que armazena o valor 10 em um espaço na memória

print(x)  # Utilizar a referência no nome atribuído
```

##Regras


**Nome da variável**
* Começar com letra (a-z, A-Z) ou sublinhado (_)
* São case sensitive, portanto: A != a

**Boas práticas para o nome da variável**

> Exemplo: Multiplicação de 2 números

*Não usar*

* Multiplicação
* MultiplicaçãoDeDoisNumeros
* m

*Usar*

* multiplicação
* mult
* mult_total (snake_case)

*Tipos de variáveis*

> Python é uma linguagem de alto nível
> Não é necessário especificar o tipo de dados quando declarar a variável
> Python define o tipo de dados na declaração

**Exemplos:**
* num = 10 (Int)
* num = 10.6 (Float)
* nome = "Rodrigo" (String)
