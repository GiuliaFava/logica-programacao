# Condicional

If

```python
if 5 < 10:
  print("verdade")
```

If / Else

```python
num1 = 10
num2 = 5
if num1 > num2:
  print("O num1 é o maior número")

else:
    print("O num2 é o maior número")
```

If / Elif / Else

```python
num1 = 100
num2 = 10
if num1 > num2:
  print("o num1 é maior que mum2")
elif num1 < num2:
  print("num1 é menor que num2")
else:
  print("Os dois números são iguais")
```
