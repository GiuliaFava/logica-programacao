# Lista
* Formada por [  ]
* Pode ser colocado qualquer tipo de informação
* Existem várias funções para utilizar com a lista
* A lista permite ser modificada (add, deletar...)
* Uma lista começa sempre pelo 0 (índice 0)

```python
list = ["Giulia", 21, 16.8]
```

```python
type (list)
```

```python
list.append("Fava")    # Adcionar um elemenrto na lista
print(list)
```

```python
print(list[0])      # Procura do índice 0 na lista
```

# Tuple
* Formada por (   )
* Não pode ser alterada
* Pode ser colocado qualquer tipo de informação
* O primeiro índice também é o 0
* Muito utilizado quando existe uma listagem de valores e constantes que não podem ser alterados

```python
tupla = ("Giulia", 21, 16.8)
```

```python
type (tupla)
```

```python
print (tupla)
print (tupla[2])
```

# Dicionário
* Formado por {   }
* Permite que você associe uma chave específica para cada valor
> ex: índice: nome; valor: Giulia
* Para adcionar um elemento não é preciso usar a função "append" basta colocar a chave do índice que deseja adcionar que será automaticamente adcionado ao final da lista
* Para fazer a procuro de um índice específico basta utilizar a chave criada para armazenar tal valor


```python
dicionario = {"Nome": "Giulia","Idade":"21","Cidade":"Lisboa"}
```

```python
type(dicionario)
```

```python
print(dicionario)
```

```python
print(dicionario["Idade"])
```

```python
dicionario["ultimo_nome"] = "Gois"
```

```python
print(dicionario)
```
