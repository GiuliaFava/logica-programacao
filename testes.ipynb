```python
# Calcular a média das nota dos alunos

notas =[85, 70, 99, 53, 95]  #Lista de notas dos alunos

#Cálculo da média das notas
soma_notas = 0
for nota in notas:
  soma_notas += nota
  media_notas = soma_notas / len(notas)

#Classificação dos alunos com base no nota mínima exigida
alunos_aprovados = []
alunos_reprovados = []

for nota in notas:
  if nota > 70:
    alunos_aprovados.append(nota)
  else:
    alunos_reprovados.append(nota)

#Imprimir os resultados
print("As notas dos alunos são: ", notas)
print("A média das notas dos alunos é: ", media_notas)
print("Os alunos aprovados foram: ", alunos_aprovados)
print("Os alunos reprovados foram: ", alunos_reprovados)

```
