**Input**


*   Armazenar a informação digitada pelo usuário em uma variável
*   O dado armazenado sempre será um teste



```python
nome = input("Digite o seu nome: ")   #Aguardar que o usuário digite
print(nome)
```

```python
nota = input("Digite a nota da sua prova: ")
print(nome, "sua nota foi", nota)
```

```python
valor = input("Digite o valor do produto:")
print(valor)
```

```python
desconto = valor - 5   #não da certo, pois o valor do produto está armazenado como str
```

```python
valor_convert = int(valor)
desconto = valor_convert - 5          #criar uma nova variável quando a mesma ser utilizada novamente no andamento do código
print(desconto)
```

```python
desconto = int(valor) - 5             #fazer a conversão direta da variável quando a mesma for somente utilizada nessa etapa
print(desconto)
```
